```python
#question 3 part 1 
#Use python to plot your normal_density function for a couple different values of µ and σ^2

import numpy as np 
import matplotlib.pyplot as plt 

mean = 0
variance = 1
x = 0 

def normal_density(mean, variance, x): #defining the normal distribution function 
    function = (1/(np.sqrt(2*np.pi*variance)))*np.exp(-(x-mean)**2/(2*variance))
    return function

print (normal_density(mean,variance,x)) 

x = np.linspace(-10, 10, 100)

y1 = normal_density(3,2,x) #different values of µ and σ^2
plt.plot(x,y1)

y2 = normal_density(2,5,x)
plt.plot(x,y2)

y3 = normal_density(1,2,x)
plt.plot(x,y3)

y4 = normal_density(0,1,x)
plt.plot(x,y4)

plt.title("Normal Density Curves for Different Mean and Variance")
plt.show()
```

```python
#question 3 part 2
#implementing numerical integration via the midpoint method with the normal distribution curve

mean = 0 
variance = 1 
a = -1
b = 5

def normal_density(mean, variance, x): #defining the normal distribution function, again 
    function = (1/(np.sqrt(2*np.pi*variance)))*np.exp(-(x-mean)**2/(2*variance)) 
    return function

def integration(mean, variance, a,b,n): #defining the midpoint method of numerical integral 
    n = 1000
    w = (b-a)/n #width of the rectangles in Riemann sum 
    midpoints = np.linspace(a+(w/2),b+(w/2),n) #creating an equally spaced between a and b by n, and adding a with midpoint 
    y = normal_density(mean,variance,midpoints) 
    integral = np.sum(y*w) #defining the numerical integral as the sum of all midpoints 
    return(integral)

print("The integral in the interval [-1,5] is",integration(mean,variance,a,b,1000)) #this value represents the probability that a normally distributed variable 'x' falls within -1≤x≤5
    
```

```python
#question 4 part c

mean = 171 #value in cm 
variance = 7.1**2 


def normal_density(mean, variance, x): #defining the normal distribution function, againx2 
    function = (1/(np.sqrt(2*np.pi*variance)))*np.exp(-(x-mean)**2/(2*variance)) 
    return function

def integration(mean, variance, a,b,n): 
    n = 1000
    w = (b-a)/n
    midpoints = np.linspace(a+(w/2),b+(w/2),n)
    y = normal_density(mean,variance,midpoints) 
    integral = np.sum(y*w) #defining the numerical integral as the sum of all midpoints 
    return(integral)

def avg_dosage(mean,variance,a,b,n):
    dosage = lambda x: 2.38*(x**2) #function for dosage D(x) = 2.38X^2
    w = (b-a)/n
    midpoints = np.linspace(a+(w/2),b+(w/2),n)
    y = normal_density(mean,variance,midpoints)
    dosage = dosage(midpoints) #defining the new function in terms of D(midpoints) 
    expected_dosage = np.sum(dosage*w*y)
    return(expected_dosage)

expected_dosage = avg_dosage(mean,variance,162,190,1000)
print("The estimated average dosage is",expected_dosage)
    
```
